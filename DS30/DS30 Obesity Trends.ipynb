# Data

```python
X = train_data.iloc[:,:-1]
y = train_data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

## PipeLine

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

#categorical_ord_pipe = Pipeline([
#    ('onehot', OrdinalEncoder(handle_unknown='use_encoded_value'))
# ])
categorical_nom_pipe = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessing = ColumnTransformer(
    [#('ord_cat', categorical_ord_pipe, cat_ord_features),
     ('nom_cat', categorical_nom_pipe, cat_nom_features)])

```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf = Pipeline([
    ('preprocess', preprocessing),
    ('classifier', RandomForestRegressor(random_state=13))
])
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
importances = rf.steps[1][1].feature_importances_
indices = np.argsort(importances)[::-1]
indices
# Print the feature ranking
print("Feature ranking:")

for f in range(X_train.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
test_rf = test_data[X_train.columns]
submit(rf.predict(test_rf), 'rf_predict')
```

3.5 Muy maal
